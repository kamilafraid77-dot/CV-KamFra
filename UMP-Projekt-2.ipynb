```python
# ðŸŽ¾ Prognozowanie wynikÃ³w turnieju tenisowego French Open Roland Garros 2025 ðŸŽ¾
### Nazwa zespoÅ‚u - MaszyniÅ›ci
### Kamila Fraid 108949
### JÄ™drzej Joniec 141443
### Piotr Palik 140595

# Zastosowany zostaÅ‚ model Gaussian Naive Bayes trenowany na chronologicznych starciach miÄ™dzy finalistami

# Wyniki:
# ZwyciÄ™Å¼czyni singla kobiet: A. Sabalenka
# ZwyciÄ™zca singla mÄ™Å¼czyzn: C. Alcaraz
```

```python
# Wymagane pakiety
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
```

```python
# 1. Dane: mecze Gauff vs Sabalenka
data = [
    # Data, Turniej, Nawierzchnia, Runda, ZwyciÄ™Å¼czyni
    ("2020-08-13", "Lexington", "Hard", "R16", "Gauff"),
    ("2020-10-22", "Ostrava", "Indoor", "R16", "Sabalenka"),
    ("2021-05-12", "Rome", "Clay", "R16", "Gauff"),
    ("2022-08-11", "Toronto", "Hard", "R16", "Gauff"),
    ("2023-03-15", "Indian Wells", "Hard", "QF", "Sabalenka"),
    ("2023-09-09", "US Open", "Hard", "Final", "Gauff"),
    ("2024-01-25", "Australian Open", "Hard", "SF", "Sabalenka"),
    ("2024-10-12", "Wuhan", "Hard", "SF", "Sabalenka"),
    ("2024-11-08", "WTA Finals", "Indoor", "SF", "Gauff"),
    ("2025-05-03", "Madrid", "Clay", "Final", "Sabalenka"),
]

# 2. DataFrame
df = pd.DataFrame(data, columns=["Date", "Tournament", "Surface", "Round", "Winner"])

# 3. Etykieta: 1 = Sabalenka wygraÅ‚a, 0 = Gauff wygraÅ‚a
df["Label"] = df["Winner"].apply(lambda x: 1 if x == "Sabalenka" else 0)

# 4. Kodowanie cech kategorycznych
le_surface = LabelEncoder()
le_round = LabelEncoder()
le_tournament = LabelEncoder()

df["Surface_enc"] = le_surface.fit_transform(df["Surface"])
df["Round_enc"] = le_round.fit_transform(df["Round"])
df["Tournament_enc"] = le_tournament.fit_transform(df["Tournament"])

# 5. ZbiÃ³r X = cechy meczu, y = etykieta (kto wygraÅ‚)
X = df[["Surface_enc", "Round_enc", "Tournament_enc"]]
y = df["Label"]

# 6. Gaussian Naive Bayes
model = GaussianNB()
model.fit(X, y)

# 7. Predykcja dla nowego meczu: Roland Garros 2025, finaÅ‚, clay
new_match = pd.DataFrame({
    "Surface": ["Clay"],
    "Round": ["Final"],
    "Tournament": ["Roland Garros"]
})

# Ponowne zakodowanie cech
new_match["Surface_enc"] = le_surface.transform(new_match["Surface"])
new_match["Round_enc"] = le_round.transform(new_match["Round"])
new_match["Tournament_enc"] = le_tournament.transform(["Madrid"]) 

X_test = new_match[["Surface_enc", "Round_enc", "Tournament_enc"]]
pred = model.predict(X_test)[0]
proba = model.predict_proba(X_test)[0]

# 8. Wynik
print("Przewidywana zwyciÄ™Å¼czyni:", "Sabalenka" if pred == 1 else "Gauff")
print(f"PrawdopodobieÅ„stwo zwyciÄ™stwa Sabalenki: {proba[1]:.2%}")
print(f"PrawdopodobieÅ„stwo zwyciÄ™stwa Gauff: {proba[0]:.2%}")
```

```python
# 1. Dane: Mecze Alcaraz vs Sinner
data = [
    # Data, Turniej, Nawierzchnia, Runda, ZwyciÄ™zca
    ("2019-04-01", "Alicante Challenger", "Clay", "R1", "Alcaraz"),
    ("2021-11-13", "ATP Masters Paris", "Indoor", "R32", "Alcaraz"),
    ("2022-07-04", "Wimbledon", "Grass", "R16", "Sinner"),
    ("2022-07-31", "Umag", "Clay", "Final", "Sinner"),
    ("2022-09-07", "US Open", "Hard", "QF", "Alcaraz"),
    ("2023-03-18", "Indian Wells", "Hard", "SF", "Alcaraz"),
    ("2023-04-01", "Miami", "Hard", "SF", "Sinner"),
    ("2023-10-04", "Beijing", "Hard", "SF", "Sinner"),
    ("2023-11-14", "ATP Finals", "Indoor", "RR", "Sinner"),
    ("2024-03-16", "Indian Wells", "Hard", "SF", "Alcaraz"),
    ("2024-06-07", "Roland Garros", "Clay", "SF", "Alcaraz"),
    ("2024-10-04", "Beijing", "Hard", "Final", "Alcaraz"),
    ("2024-12-01", "Riyadh Exhibition", "Hard", "Final", "Sinner"),  # pokazowy
    ("2025-05-18", "ATP Masters Rome", "Clay", "Final", "Alcaraz"),
]

# 2. DataFrame
df = pd.DataFrame(data, columns=["Date", "Tournament", "Surface", "Round", "Winner"])

# 3. Etykieta: 1 = Alcaraz wygraÅ‚, 0 = Sinner wygraÅ‚
df = df[df["Winner"].isin(["Sinner", "Alcaraz"])]  
df["Label"] = df["Winner"].apply(lambda x: 1 if x == "Alcaraz" else 0)

# 4. Kodowanie cech kategorycznych
le_surface = LabelEncoder()
le_round = LabelEncoder()
le_tournament = LabelEncoder()

df["Surface_enc"] = le_surface.fit_transform(df["Surface"])
df["Round_enc"] = le_round.fit_transform(df["Round"])
df["Tournament_enc"] = le_tournament.fit_transform(df["Tournament"])

# 5. ZbiÃ³r X = cechy meczu, y = etykieta (kto wygraÅ‚)
X = df[["Surface_enc", "Round_enc", "Tournament_enc"]]
y = df["Label"]

# 6. Gaussian Naive Bayes
model = GaussianNB()
model.fit(X, y)

# 7. Predykcja dla nowego meczu: Roland Garros 2025, finaÅ‚, clay
new_match = pd.DataFrame({
    "Surface": ["Clay"],
    "Round": ["Final"],
    "Tournament": ["Roland Garros"]
})

# Ponowne zakodowanie cech
new_match["Surface_enc"] = le_surface.transform(new_match["Surface"])
new_match["Round_enc"] = le_round.transform(new_match["Round"])
new_match["Tournament_enc"] = le_tournament.transform(["Roland Garros"])

X_test = new_match[["Surface_enc", "Round_enc", "Tournament_enc"]]
pred = model.predict(X_test)[0]
proba = model.predict_proba(X_test)[0]

# 8. Wynik
print("Przewidywany zwyciÄ™zca:", "Alcaraz" if pred == 1 else "Sinner")
print(f"PrawdopodobieÅ„stwo zwyciÄ™stwa Alcaraza: {proba[1]:.2%}")
print(f"PrawdopodobieÅ„stwo zwyciÄ™stwa Sinnera: {proba[0]:.2%}")
```
