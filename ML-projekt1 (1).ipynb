```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

# Oryginalna tabela
data = {
    "Nazwisko": [
        "Trzaskowski", "Nawrocki", "Mentzen", "Holownia", "Zandberg",
        "Biejat", "Braun", "Stanowski", "Bartoszewicz", "Seneszyn", "Jakubiak"
    ],
    "7-8/05/2025": [32.3, 22.7, 13.4, 8.4, 7.0, 5.5, 5.3, 2.0, None, 1.4, 0.9],
    "28-29/04/2025": [33.0, 25.0, 15.0, 7.0, 6.0, 4.0, 4.0, 2.0, 1.0, 1.0, 1.0],
    "15-16/04/2025": [34.0, 24.0, 16.0, 8.0, 5.0, 3.0, 4.0, 3.0, None, None, None],
    "31/03-1/04/2025": [35.0, 23.0, 19.0, 8.0, 4.0, 3.0, 3.0, None, None, None, None],
    "7-9/03/2025": [37.0, 22.0, 21.0, 8.0, 2.0, 3.0, 3.0, 2.0, None, None, None],
    "8-10/02/2025": [35.0, 25.0, 16.0, 8.0, 2.0, 2.0, 3.0, 5.0, None, None, 2.0],
    "1/20/2025": [37.0, 27.0, 14.0, 11.0, 2.0, 3.0, 3.0, None, None, None, 1.0],
    "10-11/12/2024": [42.0, 28.7, 12.7, 10.7, None, 3.3, None, None, None, None, 2.7]
}

df = pd.DataFrame(data)

# === 1. Przekształcenie do formatu long ===
df_long = df.set_index("Nazwisko").T
df_long.index = pd.to_datetime(df_long.index, dayfirst=True, errors='coerce')  # konwersja dat

df_long = df_long.sort_index()  # sortowanie chronologiczne

# === 2. Budowa sekwencji czasowych ===
sequence_length = 3  # ile poprzednich pomiarów użyć

X_list = []
y_list = []
name_list = []

for name in df_long.columns:
    series = df_long[name].dropna()
    for i in range(len(series) - sequence_length):
        X_list.append(series.iloc[i:i+sequence_length].values)
        y_list.append(series.iloc[i+sequence_length])
        name_list.append(name)

X = np.array(X_list)
y = np.array(y_list)
names = np.array(name_list)

# === 3. Trening modelu ===
model = XGBRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# === 4. Predykcja następnego wyniku dla każdego kandydata ===
next_preds = {}
for name in df_long.columns:
    series = df_long[name].dropna()
    if len(series) >= sequence_length:
        last_sequence = series.iloc[-sequence_length:].values.reshape(1, -1)
        pred = model.predict(last_sequence)[0]
        next_preds[name] = round(pred, 2)

# === 5. Wynik końcowy ===
results = pd.DataFrame.from_dict(next_preds, orient="index", columns=["Prognoza następnego sondażu"])
results.index.name = "Nazwisko"
results = results.reset_index().sort_values("Prognoza następnego sondażu", ascending=False)

print(results)

```
